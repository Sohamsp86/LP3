```python
def issafe(board,row,col):
    for i in range(row):
        if board[i]==col or abs(board[i]-col)==abs(i-row):
            return False
    return True


def solve(board,row):
    if row==8:
        print_board(board)
        return
    
    for i in range(8):
        if(issafe(board,row,i)):
            board[row]=i
            solve(board,row+1)
            board[row]=-1

def print_board(board):
    for i in range(8):
        row=['*']*8
        row[board[i]]='Q'
        print(' '.join(row))
    print()

board=[-1]*8
board[0]=0
solve(board,1)
```

```python
# Function to check if placing a Queen at (row, col) is safe
def issafe(board, row, col):
    # Check for any conflict with previously placed Queens
    for i in range(row):
        # Check if there's a Queen in the same column or on the same diagonal
        if board[i] == col or abs(board[i] - col) == abs(i - row):
            return False  # Not safe to place Queen here
    return True  # Safe to place Queen

# Recursive function to solve the 8-Queens problem by placing Queens row by row
def solve(board, row):
    # Base case: If 8 Queens are successfully placed, print the board
    if row == 8:
        print_board(board)  # Print the board configuration
        return
    
    # Try placing a Queen in each column of the current row
    for i in range(8):
        # Check if it's safe to place the Queen in this column
        if issafe(board, row, i):
            board[row] = i  # Place the Queen in column i of the current row
            solve(board, row + 1)  # Recur to place the Queen in the next row
            board[row] = -1  # Backtrack by removing the Queen from the current position

# Function to print the board configuration based on the board array
def print_board(board):
    for i in range(8):
        row = ['*'] * 8  # Initialize an empty row
        row[board[i]] = 'Q'  # Place the Queen at the column indicated by board[i]
        print(' '.join(row))  # Print the row with Queens and empty cells
    print()  # Print a new line after each board configuration for clarity

# Initialize the chessboard with -1, meaning no Queen is placed in any row
board = [-1] * 8

# Place the first Queen at (0, 0) as specified
board[0] = 0

# Start solving from the second row (row index 1) since the first Queen is already placed
solve(board, 1)
```
