```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('kmeans.csv', sep=',',encoding='Latin-1')
df
```

```python
features=df[['QUANTITYORDERED','PRICEEACH']]
scaledfeatures=StandardScaler().fit_transform(features)
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaledfeatures)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel('no.ofclusters')
plt.ylabel('wcss')
plt.title('elbowmethod')
```

```python
optimal=3
kmeans=KMeans(n_clusters=optimal,random_state=0)
clusterlabels=kmeans.fit_predict(scaledfeatures)
```

```python
plt.scatter(scaledfeatures[:,0],scaledfeatures[:,1],c=clusterlabels,cmap='viridis')
```
